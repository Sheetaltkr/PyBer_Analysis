```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot

plt.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)

```

```python
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
#set y limit between 0 and 45
plt.ylim(0,45)
#Create a title
plt.title("Pyber Fare by Month")
#Create plot and add label to legend
plt.plot(x_axis,y_axis,marker='D',color='blue', linewidth=2, label="Boston")
#Add the legend
plt.legend()
# Add a grid.
plt.grid()
```

```python
fig, ax = plt.subplots()
ax.invert_yaxis()
ax.scatter( y_axis,x_axis,s=[i*5 for i in y_axis],color="lightblue",edgecolors="black",label="Boston")
ax.set_xlim(0,50)
ax.legend()
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
```

```python
plt.subplots(figsize=(8, 8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis,labels=x_axis,autopct="%1.1f%%",colors=colors,explode=explode_values)
plt.show()
```

```python

#y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#Source for sort function code: https://stackoverflow.com/questions/13070461/get-indices-of-the-top-n-values-of-a-list
#--------------------------------------------------------------------------------
top2_yaxis_value_indx = sorted(range(len(y_axis)), key=lambda i: y_axis[i])[-2:]
#--------------------------------------------------------------------------------

explode_values=[]

for a in y_axis:
    explode_values.append(0)
    
for a in top2_yaxis_value_indx:
    explode_values[a]=0.2
    
explode_values
```

```python

fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis,autopct="%1.1f%%",colors=colors,explode=explode_values,shadow = True,startangle=270)
#plt.subplots(figsize=(8,8))
fig.set_size_inches(8,8)
ax.invert_yaxis()
plt.show()
```
